```python
from abc import ABC, abstractmethod

class Stock:
    counter = 0
    stock_set = set()

    def __init__(self, stock_code, desc, price, available):
        self.stock_code = stock_code
        self.desc = desc
        self.price = price
        self.available = available
        Stock.counter += 1
        Stock.stock_set.add(self.desc)

    def sell_price(self):
        sp = round(self.price, 2)
        print('cost = $', sp)

    def label(self):
        print(f"{self.desc} costs you {self.price}")

    def sale(self, discount):
        new_sp = round(self.price * (1 - discount), 2)
        print('the discounted price of {} is $ {}'.format((self.stock_code, self.desc), new_sp))
        return new_sp


class Chocolate(Stock):

    category = 'chocolate'

    def __init__(self, stock_code, desc, price, available):
        super().__init__(stock_code, desc, price, available)

    def __repr__(self):
        print("You have chosen = {self.label}")


class Chips(Stock):
    category = 'chips'

    def __init__(self, stock_code, desc, price, packet_size):
        Stock.__init__(self, stock_code, desc, price, packet_size)

    def __repr__(self):
        print("You have chosen = {self.label}")

    def one_for_one(self):
        # buy one get one offer
        sp = round(self.price, 2)
        print('Buy two packets of {} and get one free. You pay {}'.format(self.packet_size, sp))


class Canned(Stock):
    category = 'cans'

    def __init__(self, stock_code, desc, price, available):
        Stock.__init__(self, stock_code, desc, price, available)

    def label(self):
        print(f'{self.desc}')
        self.sell_price()

    def apply_discount(self, discount):
        new_sp = self.sale(discount)
        print('Price reduced ! you save = $ ', self.price - new_sp)

class VendingMachine:
    
    products = {}
    money = 0
        
    def display_vending_machine(self):
        for k, v in self.products.items():
            print(k, v)

    
class StockManipulation(ABC):
    
    @abstractmethod
    def update_stock(self, product):
        pass
    
    
class StockManipulation_Print(ABC):
    
    @abstractmethod
    def update_stock(self, product):
        pass
    
    @abstractmethod
    def print_special_attribute(self, product):
        pass
    

class MoneyDependency(ABC):
    
    @abstractmethod
    def do_you_have_enough_money(self):
        pass
    

class CashMoney(MoneyDependency, VendingMachine):
    
    def do_you_have_enough_money(self):
        return True
    
    def print_type_of_money(self):
        print("type of money = cash")

class CoinMoney(MoneyDependency, VendingMachine):    
    
    def do_you_have_enough_money(self):
        return True
    
    def print_type_of_money(self):
        print("type of money = coin")

    
class StockDetailsCan(StockManipulation_Print, VendingMachine):
    
    def __init__(self, volume):
        self.volume = volume


    def update_stock(self, product):
        d2 = {product.desc: {}}
        d2[product.desc]['price'] = product.price
        d2[product.desc]['quantity'] = product.available
        d2[product.desc]['volume'] = self.volume 
        self.products.setdefault(product.stock_code, {}).update(d2)
        
    def print_special_attribute(self):
        print("product has volume = ", self.volume)

class StockDetailsChips(StockManipulation_Print, VendingMachine):

    def __init__(self, flavour, packet_size):
        self.flavour = flavour
        self.packet_size = packet_size        
        
    def update_stock(self, product):
        d2 = {product.desc: {}}
        d2[product.desc]['price'] = product.price
        d2[product.desc]['quantity'] = product.available
        d2[product.desc]['flavour'] = self.flavour
        d2[product.desc]['packet_size'] = self.packet_size
        self.products.setdefault(product.stock_code, {}).update(d2)
        
    def print_special_attribute(self):
        print(f"product has flavour = {self.flavour } and packet_size = {self.packet_size}")

class StockDetailsChocolate(StockManipulation, VendingMachine):

        
    def update_stock(self, product):
        d2 = {product.desc: {}}
        d2[product.desc]['price'] = product.price
        d2[product.desc]['quantity'] = product.available
        self.products.setdefault(product.stock_code, {}).update(d2)


def choose_product(product):
    
    product_name = product.stock_code
    product_count = product.available
    print("$"*100)
    print(f"You have chosen {product_name} {product.desc}")
    if product_count != 0:
        print(f"feed in atleast {product.price} or more: ")
        money = float(input())
        cash_money = CashMoney()
        coin_money = CoinMoney()
        if money < 10:
            coin_money.print_type_of_money()
        else:
            cash_money.print_type_of_money()       
 
        if money > product.price:
            balance = money - product.price
            print(f"Here is the balance : {balance}")
            product.available -= 1
        elif money == product.price:
            print("Perfect Change - Thank you")
            product.available -= 1
        else:
            print("Feed in more cash/coin")
    print("$"*100)
    print("Thank you for choosing this vending machine")
    print("$"*100)

print('****************************************************************')

ca_1 = Canned('Can', 'Coke', 5.00, 500)
ca_2 = Canned('Can', 'Pepsi', 4.00, 250)
ch_1 = Chips('Chips', 'Potato', 10.00, 100)
ch_2 = Chips('Chips', 'Tapioco', 10.00, 100)
chl_1 = Chocolate('Chocolate', 'Kitkat', 1.00, 15)
chl_2 = Chocolate('Chocolate', 'Hersheys', 2.00, 20)
print("Total number of products available: ",Stock.counter)
print("products = ", Stock.stock_set)
vm = VendingMachine()
can_obj = StockDetailsCan(5)
can_obj.update_stock(ca_1)
can_obj.print_special_attribute()
can_obj.update_stock(ca_2)
can_obj.print_special_attribute()
chips_obj = StockDetailsChips('salt', 2)
chips_obj.update_stock(ch_1)
chips_obj.print_special_attribute()
chips_obj.update_stock(ch_2)
chips_obj.print_special_attribute()
choco_obj = StockDetailsChocolate()
choco_obj.update_stock(chl_1)
choco_obj.update_stock(chl_2)
vm = VendingMachine()
vm.display_vending_machine()
choose_product(ca_1)
can_obj.update_stock(ca_1)
vm.display_vending_machine()
```
